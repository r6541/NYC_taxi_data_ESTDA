####Create stratified sample using the medallion


```python
# Define the stratified sampling based on medallion
stratified_sample_medallion = trip.groupby('medallion', group_keys=False).apply(lambda x: x.sample(frac=0.01))

# Display the stratified sample
print(stratified_sample_medallion)
```

```python
stratified_sample_medallion.to_csv('/content/drive/MyDrive/Colab Notebooks/trip_sample1.csv', index=False)
```
